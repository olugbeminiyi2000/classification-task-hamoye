```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (abs(corr_matrix.iloc[i, j]) > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                
    return col_corr
```

```python
corr_features = correlation(eprice_data, 0.9)
corr_features
```

```python
 eprice_data.drop(corr_features, axis=1, inplace=True)
```

```python
eprice_data['ECPD'] = eprice_data['Global_active_power']*1000/60 - eprice_data['Sub_metering_1'] - eprice_data['Sub_metering_2'] - eprice_data['Sub_metering_3']
eprice_data
```

```python
eprice_data[ "Date" ] = pd.to_datetime(eprice_data[ "Date" ]) #convert column to datetime
eprice_data.set_index( "Date" , inplace= True )
```

```python
eprice_data
```

```python
#now to shrink our data to perday so we could get the price spent on energy per day
eprice_ddata = eprice_data.resample( 'D' ).mean()
print(eprice_ddata.isnull().values.any())
eprice_ddata.interpolate(method= 'time' , limit_direction= 'forward' , inplace=True)

```

```python
eprice_ddata
#now we know the average of everything used for a day now to calculate the electrical power spent 
#each day and after research true power is also known as active power it is the power that does work
#reactive power is power released and stored in capacitors without being consumed and it returns to its source
#so to get electrical price per day we need the true power which could also have been gotten by the multiplication
#of the (voltage and global_intensity) but the global_intensity and global_active_power are highly correlated together
#which could affect our forecast so I decided to drop the global_intensity since I already have power needed to get
#electrical price per day all i could be a very important predictor to knowing future electrical prices .
```

```python
#now we have the energy consumed per day with the price multiplied with 'ECPD' gives the electric price per day 
#but we dont need the price since all observation will be multiplied with the same amount we dont need it here.
eprice_ddata = eprice_ddata.reset_index()
eprice_ddata
```

```python
from fbprophet import Prophet
X_train[ 'Date' ]= pd.to_datetime(X_train[ 'Date' ])
X_train = X_train.rename(columns={ "Date" : "ds" , "ECPD" : "y" })
X_train.head(40) 
```

```python
X_train = X_train.iloc[:,[0,7]]
X_train.head(5)
```

```python
model = Prophet()
model.fit(X_train)
#predict for the next 30 days
future = model.make_future_dataframe(periods= 30 , freq= 'D' )
forecast = model.predict(future)
#here after we have fitted our train_data in the model this model predicts it's own train data and we can also make it do further prediction
#by using make_future_dataframe method the value and the frquency either daily,seconds,minutes,months,yearly and so on
#then we forecast all i.e the xtrain and the future data
```

```python
model.plot(forecast)
plt.show() 
```

```python
y_true = X_train.iloc[:,[1]].values
y_true
```

```python
y_pred = forecast.iloc[:1019,[18]].values
y_pred
```

```python
#evaluation metrics 
#mean absolute error
#mean absolute percentage error
#root mean square error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_true,y_pred)
rmse = np.sqrt(mean_squared_error(y_true,y_pred))
mape = mae/sum(y_true) * 100
r2_score = r2_score(y_true, y_pred)
```
